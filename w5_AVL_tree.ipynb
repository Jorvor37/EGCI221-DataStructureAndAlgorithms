# Single Rotation

```python
def height(node):
    if node is None:
        return -1
    return node.height

def rotateRight(k2): # 1. k2 point to the root node
    """
    1. k2 point to the root node
    2. Assign pointer k1 to left child of k2
    3. change k2 left child to k1 right child
    4. k1 right child to k2
    5. update height of this 2 sub trees
    """
    k1 = k2.left            # 2. Assign pointer k1 to left child of k2
    k2.left = k1.right      # 3. change k2 left child to k1 right child
    k1.right = k2           # 4. k1 right child to k2
    k2.height = max(height(k2.left), height(k2.right)) + 1  # 5. update height of k2(right-side & down)
    k1.height = max(height(k1.left), height(k1.right)) + 1  # 5. update height of k1(left-side & up)
    return k1

def rotateLeft(k1): # 1. k1 point to the root node
    """
    1. k1 point to the root node
    2. Assign pointer k2 to right child of k1
    3. change k1 right child to k2 left child (change the pointer)
    4. k2 left child to k1
    5. update the hight of this 2 sub trees
    """

    k2 = k1.right           # 2. Assign pointer k2 to right child of k1
    k1.right = k2.left      # 3. change k1 right child to k2 left child (change the pointer)
    k2.left = k1            # 4. k2 left child to k1
    k1.height = max(height(k1.left), height(k1.right)) + 1  # 5. update height of k1(left-side & down)
    k2.height = max(height(k2.left), height(k2.right)) + 1  # 5. update height of k2(right-side & up)
    return k2
```

# Double Rotation

```python
def rotateLR(k3):
    """
    1. k3 point to the root node
    2. Said that left of root node to rotate left
    3. Say that rotate k3 right again
    """
    k3.left = rotateLeft(k3.left)
    return rotateRight(k3)

def rotateRL(k3):
    k3.right = rotateRight(k3.right)
    return rotateLeft(k3)
```
